```python
import pandas as pd
df = pd.read_csv("age1.csv")
df.head(10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Age'],df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```
