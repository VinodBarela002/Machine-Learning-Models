```python
import pandas as pd
```

```python
dataset = pd.read_csv('car.csv')
dataset.head()
```

```python
x = dataset[['Age','EstimatedSalary']]
y = dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.29,random_state=0)
```
