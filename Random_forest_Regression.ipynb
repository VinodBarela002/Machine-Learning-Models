```python
import pandas as pd
```

```python
dataset = pd.read_csv('age1.csv')
X = dataset[['Age']]
Y = dataset[['Income']]
dataset
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X, Y)
```

```python
regressor.predict([[70]])
```

```python
regressor.predict(X)
```

```python
regressor.score(X,Y)
```
